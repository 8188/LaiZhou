```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import time
import os
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams["axes.unicode_minus"] = False
```

```python
%%time
df = pd.read_pickle('C:/Users/jsy19/OneDrive - 薅羊毛大学/文档/mystuff/data/莱州/pre_data_new.pkl')
```

```python
tar_fe = [
    '发电机内冷水箱氢气泄漏', '发电机密封油励侧回油氢气泄漏', '发电机密封油汽侧回油氢气泄漏', 
    '发电机封闭母线A相氢气泄漏', '发电机封闭母线B相氢气泄漏', '发电机封闭母线C相氢气泄漏', 
    '发电机封闭母线中性点1氢气泄漏', '发电机封闭母线中性点2氢气泄漏',
    '进氢压力',
    '发电机内氢气纯度',  '发电机出氢湿度', 
    '发电机密封油-氢气差压', '发电机密封油油含水量', 
    '发电机定子冷却水电导率', '发电机定子冷却水油水PH计', '发电机定子冷却水流量1差压'
]
tar_unit = [
    '%', '%', '%',
    '%', '%', '%',
    '%', '%',
    'MPa',
    '%', '℃',
    'KPa', '%',
    'μS/cm', '',
    'KPa'
]
tar_alarm = [
    4, 4, 4,
    4, 4, 4,
    4, 4,
    0.48,
    90, 0,
    [36, 76], 10,
    0.5, [7, 9], 20
]
dic = dict(zip(tar_fe, tar_alarm))
tingji = df.query('发电机频率<49.5 | 发电机频率>50.5')
```

```python
fig, axes = plt.subplots(nrows=len(tar_fe)//2, ncols=2, figsize=(15, 40), dpi=100)
for i, col in enumerate(tar_fe):
    plt.subplot(len(tar_fe)//2, 2, i+1)
    plt.plot(df['time'], df[col], alpha=0.5)
    plt.scatter(tingji['time'], tingji[col], c='green', s=1, label='停机')
    plt.axhline(df[col].mean(), ls='-', lw=1, c='black', label='均值', alpha=0.5)
    plt.axhline(df[col].median(), ls='-', lw=1, c='yellow', label='中位数', alpha=0.5)
    if isinstance(dic[col], list):
        plt.axhline(y=dic[col][0], ls='dashed', c='red', label='低报警')
        plt.axhline(y=dic[col][1], ls='dashed', c='red', label='高报警')
    else:
        plt.axhline(y=dic[col], ls='dashed', c='red', label='报警值')
    plt.ylabel(tar_unit[i])
    plt.title(col, fontsize=14)
    plt.legend()
```
